```python
from langchain.llms.openai import OpenAI
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate, ChatPromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler

chat = ChatOpenAI(
    temperature=1,
    model_name="gpt-5-nano",
    streaming=True,
    callbacks=[StreamingStdOutCallbackHandler()]
    )

chef_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a world-class international chef. You create easy to follow "
    "recipies for any type of cuisine with easy to find ingredients"),
    ("human", "I wnat to cook {cuisine} food")
])

chef_chain = chef_prompt | chat

```

```python
veg_chef_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a vegetarian chef specialized on making traditional recipies"
    "vegetarian. You find alternative ingredients and explain their preparation. You"
    "don't radically modify the recipe. If there is no"
    "alternative for a food just say you don't know how to replace it."),
("human","{recipe}")
])

veg_chain = veg_chef_prompt | chat

final_chain = {"recipe":chef_chain} | veg_chain

final_chain.invoke({
    "cuisine":"indian"
})
```

```python
chat_prompt2 = ChatPromptTemplate.from_messages([
   ("system", 
     "You are a world-class pianist. When you hear a composer's name, "
     "you can teach all of that composer's pieces."),
    ("human", "Tell me about {composer}")
])

chain = chat_prompt2 | chat

basic_prompt = ChatPromptTemplate.from_messages([
    ("system", 
     "You are excellent at teaching piano beginners. When a beginner wants to learn "
     "a composer's works, you will select and recommend the most suitable piece "
     "from that composer's repertoire."),
    ("human", "{pieces_list}")
])

basic_chain = basic_prompt | chat

final_chain = {"pieces_list":chain} | basic_chain

final_chain.invoke({
    "composer":"Bach"
})
```
